# Importing Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#EDA

```python
dataset =pd .read_csv('train.csv')
dataset1=pd.read_csv('gender_submission.csv')
dataset2=pd.read_csv('test.csv')
print('dataset head ',dataset.head())
print('information',dataset.info())
print('describe',dataset.describe())
```

# Data Cleaning

```python
dataset.drop(['Ticket'],axis=1,inplace=True)
dataset.drop(['PassengerId'],axis=1,inplace=True)
dataset.drop(['Cabin'],axis=1,inplace=True)
dataset.drop(['Name'],axis=1,inplace=True)
```

```python
print(dataset.isnull().sum())
```

```python
dataset['Age'].fillna(dataset['Age'].median(),inplace=True)
```

```python
dataset['Embarked'].fillna(dataset['Embarked'].mode()[0],inplace=True)
```

```python
dataset['Fare'].fillna(dataset['Fare'].median(),inplace=True)
```

```python
print(dataset.isnull().sum())
```

# Encoded Categorical Variables:

```python
print(dataset.head())
```

```python
dataset['Embarked']=dataset['Embarked'].map({'C':0,'Q':1,'S':2})
dataset['Sex']=dataset['Sex'].map({'male': 0,'female':1})
```

```python
print(dataset.head())
```

# splitting The  Dataset

```python
x=dataset.drop(['Survived'],axis=1)
y=dataset['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
dataset=scaler.fit_transform(dataset)

```

# Model building

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression()
model2.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier(n_estimators=50,max_depth=10,min_samples_leaf=1,min_samples_split=10,criterion='gini',random_state=0,class_weight='balanced')
model3.fit(x_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model4=KNeighborsClassifier(n_neighbors=5)
model4.fit(x_train,y_train)
```

```python
from sklearn.svm import SVC
model5=SVC(kernel='rbf',random_state=0,C=100,gamma='scale')
model5.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
model6=DecisionTreeClassifier(criterion='gini',max_depth=10,random_state=0)
model6.fit(x_train,y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model1=GradientBoostingClassifier(learning_rate= 0.01, max_depth= 3, min_samples_split= 2, n_estimators= 50, random_state=0,subsample= 0.8)
model1.fit(x_train,y_train)
#
```

# classification report


```python
from sklearn.metrics  import accuracy_score,classification_report,confusion_matrix
models={
    model1:'GradientBoostingClassifier',
    model2:'LogisticRegression',
    model3:'RandomForestClassifier',
    model4:'KNeighborsClassifier',
    model5:'SVC',
    model6:'DecisionTreeClassifier'
}
for model, model_name in models.items():
    y_pred = model.predict(x_test)
    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print("-" * 50)
```
